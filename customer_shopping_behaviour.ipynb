### Imputing missing values in Review Rating column with the median rating of the product category

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# Checking Null Value Again 
df.isnull().sum()
```

### Renaming columns according to snake casing for better readability and documentation

#### Better Readability – Lowercase and snake_case make column names clean and consistent.

#### Easy to Remember – Simple names like purchase_amount are quicker to type and reduce mistakes.

#### Avoid Errors – Removing spaces and special characters prevents syntax issues in coding and SQL.

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

### create a new column age_group


#### Data Segmentation – Divided age into 4 groups to better analyze customer categories.

#### Better Insights – Helps compare shopping behavior across different age groups.

#### Improved Analysis – Makes visualization and reporting more structured and meaningful.

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

### Create New Column Purchase_frequecy_days

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

### Why we DROP promo_code_used column

#### Redundant Information – promo_code_used had the same values as discount_applied, so it was duplicate data.

#### Avoid Multicollinearity – Keeping identical columns can affect analysis and machine learning models.

#### Cleaner Dataset – Removing unnecessary columns makes the dataset more efficient and easier to manage.
